## Linear regression model

Linear regression consists of finding the best-fitting straight line through the points. The best-fitting line is called a regression line.  



```python
%matplotlib inline
import numpy as np
import pylab
from IPython.display import Image
#
# rooms, distance to downtown
features = np.array([
    [4, 1],
    [1, 0.5],
    [4, 0.5],
    [2, 3],
    [4, 3],
    [3, 3]
])

y = [
    4700,
    2000,
    4700,
    3000,
    4600,
    3600,
]

X = features[:,0]
X = X.reshape(-1, 1)
pylab.scatter(X, y)
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
clf.fit(X, y)

pylab.scatter(X, y)
pylab.plot(X, clf.predict(X))
```

```python
def draw_outliner(y, showLine=False):
    y_backup = y[:]
    try:
        pylab.figure(figsize=(20,10))
        y[4] = 8000
        pylab.scatter(X, y)

        clf = linear_model.LinearRegression()
        clf.fit(X, y)
        pylab.scatter(X, y, s=200)
        if showLine:
            pylab.plot(X, clf.predict(X)  * 1.1) 
    finally:
        y = y_backup[:]

draw_outliner(y)
```

```python
draw_outliner(y, True)
```
